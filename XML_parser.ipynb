```python
import xml.etree.ElementTree as ET

import pandas as pd



# Define a function to parse the XML and flatten it into a dictionary

def parse_xml(xml_file):

    tree = ET.parse(xml_file)

    root = tree.getroot()

    data = {}



    # Iterate over each child element of the root

    for child in root:

        if child.tag == 'CUSTOMER' or child.tag == 'CONTACT' or child.tag == 'MACHINE':

            for subchild in child:

                data[subchild.tag] = subchild.text

        elif child.tag == 'SURVEYRESULT':

            for subchild in child.findall('.//b'):

                question = subchild.text.strip()

                answer = subchild.find('following-sibling::td[1]').text.strip()

                data[question] = answer



    return data



# Path to the XML file

xml_file = '/content/formatted.xml'



# Parse the XML and flatten it into a dictionary

data = parse_xml(xml_file)



# Convert the dictionary into a DataFrame

df = pd.DataFrame([data])



# Print the DataFrame

print(df)
```
